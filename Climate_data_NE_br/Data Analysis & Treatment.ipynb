```python
import pandas as pd
import seaborn as sns
import statistics  as sts
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('northeast.csv')
dataset.head(5)
```

```python
dataset.describe()
```

```python
dataset.columns
```

```python
v_dataset = dataset[(dataset['PRECIPITAÇÃO TOTAL, HORÁRIO (mm)'] > -9999.0) 
                   & (dataset['PRESSAO ATMOSFERICA AO NIVEL DA ESTACAO, HORARIA (mB)'] > -9999.0)
                   & (dataset['PRESSÃO ATMOSFERICA MAX.NA HORA ANT. (AUT) (mB)'] > -9999.0)
                   & (dataset['PRESSÃO ATMOSFERICA MIN. NA HORA ANT. (AUT) (mB)'] > -9999.0)
                   & (dataset['RADIACAO GLOBAL (Kj/m²)'] > -9999.0)
                   & (dataset['TEMPERATURA DO AR - BULBO SECO, HORARIA (°C)'] > -9999.0)
                   & (dataset['TEMPERATURA DO PONTO DE ORVALHO (°C)'] > -9999.0)]
#v_dataset = dataset[dataset['PRESSAO ATMOSFERICA AO NIVEL DA ESTACAO, HORARIA (mB)'] > -9999.0]

v_dataset.head(5)
```

```python
v_dataset.describe()
```

```python
f_dataset = v_dataset[(v_dataset['TEMPERATURA MÁXIMA NA HORA ANT. (AUT) (°C)'] > -9999.0) 
                   & (v_dataset['TEMPERATURA MÍNIMA NA HORA ANT. (AUT) (°C)'] > -9999.0)
                   & (v_dataset['TEMPERATURA ORVALHO MAX. NA HORA ANT. (AUT) (°C)'] > -9999.0)
                   & (v_dataset['TEMPERATURA ORVALHO MIN. NA HORA ANT. (AUT) (°C)'] > -9999.0)
                   & (v_dataset['UMIDADE REL. MAX. NA HORA ANT. (AUT) (%)'] > -9999.0)
                   & (v_dataset['UMIDADE REL. MIN. NA HORA ANT. (AUT) (%)'] > -9999.0)
                   & (v_dataset['UMIDADE RELATIVA DO AR, HORARIA (%)'] > -9999.0)
                   & (v_dataset['VENTO, DIREÇÃO HORARIA (gr) (° (gr))'] > -9999.0)
                   & (v_dataset['VENTO, RAJADA MAXIMA (m/s)'] > -9999.0)
                   & (v_dataset['VENTO, VELOCIDADE HORARIA (m/s)'] > -9999.0)
                   & (v_dataset['height'] > -9999.0)]
```

```python
f_dataset.head(5)
```

```python
f_dataset.describe()
```
